##5-1: Money Change Again

```python
#include <bits/stdc++.h>

using namespace std;

int solve(int money){
    vector<int> array(money + 1);
    array[1] = 1;
    array[2] = 2;
    array[3] = 1;
    array[4] = 1;
    for(int i = 5; i <= money; i++){
        int value = min(array[i - 4] + 1, array[i - 3] + 1);
        value = min(value, array[i - 1] + 1);
        array[i] = value;
    }
    return array[money];
}

int main(){
    int money;
    cin >> money;
    cout << solve(money);
    return 0;
}
```

##5-2: Primitive Calculator

```python
#include <bits/stdc++.h>

using namespace std;

void solve(int n){
    vector<int> array(n + 1);
    array[1] = 0;
    array[2] = 1;
    array[3] = 1;
    for(int i = 4; i <= n; i++){
        int value = 10e6;
        if(i % 3 == 0){
            value = min(value, array[i / 3] + 1);
        }
        if(i % 2 == 0){
            value = min(value, array[i / 2] + 1);
        }
        array[i] = min(value, array[i - 1] + 1);
    }

    cout << array[n] << endl;
    vector<int> steps(array[n] + 1);
    int i = array[n] + 1;
    steps[i--] = n;
    while(n > 0){
        steps[i--] = n;
        if(n % 3 == 0 && array[n] == array[n/3] + 1){
            n = n /  3;
        }else if(n % 2 == 0 && array[n] == array[n/2] + 1){
            n = n / 2;
        }else{
            n = n - 1;
        }
    }
    for(int i = 0; i < steps.size(); i++){
        cout << steps[i] << ' ';
    }
    cout << endl;
}

int main(){
    int n;
    cin >> n;
    solve(n);
    return 0;
}
```

##5-3: Edit Distance

```python
#include <bits/stdc++.h>

using namespace std;

int edit_distance(string A, string B, int i, int j){
    int matrix[i][j];

    for(int n = 0; n < i; n++){
        matrix[n][0] = n;
    }

    for(int m = 0; m < j; m++){
        matrix[0][m] = m;
    }

    for(int n = 1; n < i; n++){
        for(int m = 1; m < j; m++){
            if(A[n - 1] == B[m - 1]){
                matrix[n][m] = min(min(matrix[n][m - 1] + 1, matrix[n - 1][m] + 1), matrix[n - 1][m - 1]);
            }else{
                matrix[n][m] = min(min(matrix[n][m - 1] + 1, matrix[n - 1][m] + 1), matrix[n - 1][m - 1] + 1);
            }
        }
    }
    return matrix[i - 1][j - 1];
}

int main(){
    string A;
    cin >> A;
    string B;
    cin >> B;
    cout << edit_distance(A, B, A.size() + 1, B.size() + 1) << endl;
    /*edit_distance(string("EDITING"), string("DISTANCE"), 7 + 1, 8 + 1);*/
    return 0;
}
```

##5-4: Longest Common Subsequence of Two Sequences


```python
#Uses python3

import sys

def lcs2(a, b):
  #write your code here
  m = len(a)
  n = len(b)
  lenght = [[None]*(n + 1) for i in range(m + 1)]
  for i in range(m + 1):
    for j in range(n + 1):
      if i == 0 or j == 0 :
          lenght[i][j] = 0
      elif a[i-1] == b[j-1]:
          lenght[i][j] = lenght[i-1][j-1]+1
      else:
          lenght[i][j] = max(lenght[i-1][j], lenght[i][j-1])
  return lenght[m][n]

if __name__ == '__main__':
    input = sys.stdin.read()
    data = list(map(int, input.split()))

    n = data[0]
    data = data[1:]
    a = data[:n]

    data = data[n:]
    m = data[0]
    data = data[1:]
    b = data[:m]

    print(lcs2(a, b))
```
