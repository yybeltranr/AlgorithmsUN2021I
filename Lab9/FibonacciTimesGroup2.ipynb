# Bubble Sort distribution - Fibonacci time


##Fibonacci Times

###Recursive Fibonacci

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from time import process_time
```

```python
def Recursive_Fibonacci(n):
  if n == 0:
    return 0
  if n == 1:
    return 1
  return Recursive_Fibonacci(n - 1) + Recursive_Fibonacci(n - 2)
```

Calculating the time

```python
n = 50
tiempo = []
initial_time = process_time()
for i in range(0, n+1):
  inicio = process_time() 
  Recursive_F = Recursive_Fibonacci(i)
  fin =process_time()
  tiempo.append(fin - inicio) 
final_time = process_time() 
print("n = {0}, F({0}) = {1} time = {2} seconds".format(n, Recursive_F, final_time - initial_time))
```

```python
plt.figure(figsize = (10,6))
plt.plot(list(range(n+1)), tiempo)
plt.title("Fibonacci Time", fontsize = 20)
plt.ylabel("Time (s)")
plt.xlabel("Fibonacci Serie")
```

### Linear Iterative Fibonacci

```python
def Linear_Fibonacci(n):
  if n == 0:
    return 0
  a, b = 0, 1
  for i in range(1, n):
    a, b = b, a + b
  return b
```

Calculating the time

```python
from time import process_time

n1 = 1000
Linear_F = []
t1 = []

for i in range(0, n):
  initial_time1 = process_time() 
  Linear_F.append(Linear_Fibonacci(i))
  final_time1 = process_time() 
  t1.append(final_time1 - initial_time1)
print("n = {0}, F({0}) = {1} \nTotal time = {2} seconds".format(n1, Linear_F, final_time1 - initial_time1))
```

```python
plt.figure(figsize = (10,6))
plt.title('Linear Fibonacci', fontsize = 20) 
plt.plot(Linear_F, t1, label= "Fibonacci", linewidth = 2, linestyle = "--", color = "indigo")
plt.legend()  
plt.xlabel("Fibonacci Numbers")
plt.ylabel("Time in seconds")                            
plt.show()
```

### Logarithmic Iterative Fibonacci (Dynamic programming)


```python
def Logarithmic_Fibonacci(n):
    memory = {}
    def F(n):
        if n == 0:
            return 0
        if n == 1:
            return 1
        if (n - 2) not in memory:
            memory[n - 2] = F(n - 2)
        if (n - 1) not in memory:
            memory[n - 1] = F(n - 1)
        return memory[n - 1] + memory[n - 2]
    return F(n)
```

Calculating the time

```python
from time import process_time

n2 = 1000
Logarithmic_F = []
t2 = []

for i in range(0, n2):
  initial_time2 = process_time() 
  Logarithmic_F.append(Logarithmic_Fibonacci(i))
  final_time2 = process_time() 
  t2.append(final_time2 - initial_time2)
print("n = {0}, F({0}) = {1} \nTotal time = {2} seconds".format(n2, Logarithmic_F, final_time2 - initial_time2))
```

```python
plt.figure(figsize = (10,6))
plt.title('Logarithmic Fibonacci', fontsize = 20) 
plt.plot(Logarithmic_F, t2, label= "Fibonacci", linewidth = 2, linestyle = "--", color = "indigo")
plt.legend()  
#plt.xlim(0, n1) 
plt.xlabel("Fibonacci Numbers")
plt.ylabel("Time  in seconds")                          
plt.show()
```
