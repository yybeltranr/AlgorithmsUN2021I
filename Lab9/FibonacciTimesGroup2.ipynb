```python
def Recursive_Fibonacci(n):
  if n == 0:
    return 0
  if n == 1:
    return 1
  return Recursive_Fibonacci(n - 1) + Recursive_Fibonacci(n - 2)
```

### Linear Iterative Fibonacci

```python
def Linear_Fibonacci(n):
  if n == 0:
    return 0
  a, b = 0, 1
  for i in range(1, n):
    a, b = b, a + b
  return b
```

Calculating the time

```python
from time import process_time

n = 50
Linear_F = []
t = []

for i in range(0, n):
  initial_time = process_time() 
  Linear_F.append(Linear_Fibonacci(i))
  final_time = process_time() 
  t.append(final_time - initial_time)
print("n = {0}, F({0}) = {1} \ntime = {2} seconds".format(n, Linear_F, t))
```

```python
plt.figure(figsize = (10,6))
plt.title('Linear Fibonacci', fontsize = 20) 
plt.plot(Linear_F, t, label= "Fibonacci", linewidth = 2, linestyle = "--", color = "indigo")
plt.legend()  
#plt.xlim(0, n)
plt.xlabel("Fibonacci Numbers")
plt.ylabel("Time")                            
plt.show()
```

### Logarithmic Iterative Fibonacci (Dynamic programming)


```python
def Logarithmic_Fibonacci(n):
    memory = {}
    def F(n):
        if n == 0:
            return 0
        if n == 1:
            return 1
        if (n - 2) not in memory:
            memory[n - 2] = F(n - 2)
        if (n - 1) not in memory:
            memory[n - 1] = F(n - 1)
        return memory[n - 1] + memory[n - 2]
    return F(n)
```

Calculating the time

```python
from time import process_time

n1 = 1000
Logarithmic_F = []
t1 = []

for i in range(0, n1):
  initial_time = process_time() 
  Logarithmic_F.append(Logarithmic_Fibonacci(i))
  final_time = process_time() 
  t1.append(final_time - initial_time)
print("n = {0}, F({0}) = {1} \ntime = {2} seconds".format(n, Logarithmic_F, t1))
```

```python
plt.figure(figsize = (10,6))
plt.title('Logarithmic Fibonacci', fontsize = 20) 
plt.plot(Logarithmic_F, t1, label= "Fibonacci", linewidth = 2, linestyle = "--", color = "indigo")
plt.legend()  
#plt.xlim(0, n1) 
plt.xlabel("Fibonacci Numbers")
plt.ylabel("Time")                          
plt.show()
```
